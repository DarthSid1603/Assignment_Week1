## Iris Dataset

### Q1: Print the first 5 elements of your DataFrame using the command head(). How many features are there and what are their types (e.g., numeric, nominal)? Compute and display summary statistics for each feature available in the dataset. These must include the minimum value, maximum value, mean, range, standard deviation, variance, count, and 25:50:75% percentiles. 

```python
import pandas as pd
df=pd.read_csv('iris.csv')
df.head(5)
```

**SepalLength** is numeric and ratio type  
**SepalWidth** is numeric and ratio type  
**PetalLength** is numeric and ratio type  
**PetalWidth** is numeric and ratio type  
**Name** is categorical and nominal type  

```python
df['SepalLength']=df['SepalLength'].astype('float64')
df['SepalWidth']=df['SepalWidth'].astype('float64')
df['PetalLength']=df['PetalLength'].astype('float64')
df['PetalWidth']=df['PetalWidth'].astype('float64')
df_temp=pd.DataFrame({'Col_name':[],'Min':[],'Max':[],'Mean':[],'Range':[],'StdDev':[],'25 percentile':[] \
                      ,'Median':[],'75 percentile':[]})
for col in df.columns:
    if(col=='Name'):
        break
    Min=df[col].min()
    Max=df[col].max()
    Mean=df[col].mean()
    Range=Max-Min
    StdDev=df[col].std()
    per_25=df[col].quantile(0.25)
    Median=df[col].quantile(0.5)
    per_75=df[col].quantile(0.75)
    x=pd.DataFrame({'Col_name':[col],'Min':[Min],'Max':[Max],'Mean':[Mean],'Range':[Range] \
                    ,'StdDev':[StdDev],'25 percentile':[per_25],'Median':[Median],'75 percentile':[per_75]})
    df_temp=df_temp.append(x)
df_temp
```

### Q2: To illustrate the feature distributions, create a histogram for each feature in the dataset.

```python
for col in df.columns:
    if(col=='Name'):
        break
    x=df[col]
    plt.hist(x,edgecolor='k')
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.title(('Distribution of '+col), fontsize=16)
    plt.show()
```

### Q3: Create a boxplot of each feature on the same plot

```python
x=[df['SepalLength'],df['SepalWidth'],df['PetalLength'],df['PetalWidth']]
labels=['SepalLength','SepalWidth','PetalLength','PetalWidth']
colours=['blue','yellow','green','white']
bp=plt.boxplot(x, labels=labels, patch_artist=True)
for box,color in zip(bp['boxes'],colours):
    box.set( color='black', linewidth=2)
    box.set_facecolor(color)
plt.ylabel('frequency',fontsize=14)
plt.title('Boxplots of all features in iris.csv',fontsize=16)
plt.show()
```
